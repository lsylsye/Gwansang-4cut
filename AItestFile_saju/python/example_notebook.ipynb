## 4. 관상 분석 예제 (TypeScript 코드와 동일한 방식)

```python
# 관상 분석용 시스템 프롬프트 (예시)
system_prompt = """
당신은 관상 분석 전문가입니다.
제공된 얼굴 측정 데이터를 바탕으로 정확하고 상세한 관상 분석을 수행하세요.
"""

# 사용자 프롬프트 (메트릭 데이터 포함)
user_prompt = """
다음 얼굴 측정 데이터를 분석해주세요:

## 얼굴 측정 데이터
- face_ratio_wh: 0.78
- head_roll: 2.5°
- 삼정: 상=좋음, 중=보통, 하=좋음

성격, 재물운, 직업운에 대해 분석해주세요.
"""

result = call_gms_api(system_prompt, user_prompt, model="gpt-5-mini", timeout=300)
print("=" * 55)
print("📜 관상 분석 결과")
print("=" * 55)
print(result)
```

## 5. 사주 분석 예제

```python
# 사주 분석용 시스템 프롬프트 (예시)
system_prompt = """
당신은 사주명리 전문가입니다.
제공된 사주 정보를 바탕으로 정확하고 상세한 사주 분석을 수행하세요.
"""

# 사용자 프롬프트 (사주 데이터 포함)
user_prompt = """
다음 사주 정보를 분석해주세요:

## 사주 4주
- 연주: 갑자
- 월주: 갑술
- 일주: 을묘 ⭐ 일간
- 시주: 정해

재물운과 직업운에 대해 분석해주세요.
"""

result = call_gms_api(system_prompt, user_prompt, model="gpt-5-mini", timeout=300)
print("=" * 55)
print("📜 사주 분석 결과")
print("=" * 55)
print(result)
```

## 6. 배치 처리 예제 (여러 요청)

```python
import time

# 여러 프롬프트를 순차적으로 처리
prompts = [
    ("시스템 프롬프트 1", "사용자 프롬프트 1"),
    ("시스템 프롬프트 2", "사용자 프롬프트 2"),
    ("시스템 프롬프트 3", "사용자 프롬프트 3"),
]

results = []
for i, (sys_prompt, usr_prompt) in enumerate(prompts, 1):
    print(f"\n[{i}/{len(prompts)}] 처리 중...")
    start_time = time.time()
    
    try:
        result = call_gms_api(sys_prompt, usr_prompt)
        elapsed = time.time() - start_time
        results.append(result)
        print(f"✅ 완료 (소요 시간: {elapsed:.2f}초)")
    except Exception as e:
        print(f"❌ 오류: {e}")
        results.append(None)

print(f"\n총 {len(results)}개 요청 완료")
```

## 7. TypeScript 코드와 연동하기

TypeScript 코드에서 생성한 프롬프트를 Python으로 전달하는 방법:

```python
import json

# TypeScript에서 JSON 파일로 저장한 프롬프트 읽기
# 예: TypeScript에서 다음과 같이 저장했다고 가정
# fs.writeFileSync('prompts.json', JSON.stringify({systemPrompt, userPrompt}))

try:
    with open('../prompts.json', 'r', encoding='utf-8') as f:
        prompts = json.load(f)
    
    system_prompt = prompts['systemPrompt']
    user_prompt = prompts['userPrompt']
    
    result = call_gms_api(system_prompt, user_prompt)
    print(result)
except FileNotFoundError:
    print("프롬프트 파일을 찾을 수 없습니다.")
```
